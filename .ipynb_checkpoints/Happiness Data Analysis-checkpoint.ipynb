### Feature Understanding
Top and bottom happiest countries;
Happiest score density and distribution;
Factor densities;

```python
#Creating top 10 and bottom 10 data frames and concatinating them
top10 = happy_df.set_index('country_name')['happiness_score'].nlargest(10).to_frame()
bottom10 = happy_df.set_index('country_name')['happiness_score'].nsmallest(10).to_frame()
dfconcat = pd.concat([top10, bottom10], axis=0)
dfconcat = dfconcat.sort_values(by="happiness_score")

fig = plt.figure(figsize=(10,8))
cmap = plt.cm.RdYlGn
norm = plt.Normalize(vmin=dfconcat['happiness_score'].min(), vmax=dfconcat['happiness_score'].max())
colors = [cmap(norm(value)) for value in dfconcat['happiness_score']]

g = plt.barh(y=dfconcat.index, width=dfconcat["happiness_score"], color=colors)
plt.ylabel("Country")
plt.xlabel("Happiness Score")
plt.title("10 happiest and least 10 happiest Countries", fontsize = 20)

plt.show()
```

```python
# Look at the distribution of happiness scores, plotting mean (red) and median (black) lines

plt.figure(figsize=(10,6))
plt.title("Happiness score distribution of the countries", fontsize = 20)
plt.xlabel('Happiness score')
plt.ylabel('Number of countries')
plt.hist(happy_df.happiness_score, bins = np.arange(0,10,0.5), edgecolor = 'white')

plt.axvline(x = happy_df['happiness_score'].mean(), color = 'red')
plt.axvline(x = happy_df['happiness_score'].median(), color = 'black')
```

We can observe the mean (red line) and median (black line) values are similar, indicating the happiness score has quite a symmetrical distribution with a slight left skew. The spread is significant and a couple of very low happiness scores are present, isolated somewhat at the low happines score scale. From above, we know these to be the scores of Afghanistan and Lebanon.

### Mean Happiness Score By Region

```python
# lets grab the mean happiness score by each region

mean_happiness = happy_df.groupby('regional_indicator')[['happiness_score']].mean()
mean_happiness = mean_happiness.sort_values(by=['happiness_score'], ascending=False)
mean_happiness
```

```python
# getting the mean of all other regions aside Europe to test the hypothesis
data = {
    'regional_indicator': ['Oceania', 'North America', 'South America', 'Asia', 'Africa', 'Europe'],
    'happiness_score': [7.109000, 6.391310, 6.251167, 5.832500, 5.277618, 4.419486]
}
happy_df = pd.DataFrame(data)

# Filter data for the specific months
regions_of_interest = ['Oceania', 'North America', 'Asia', 'South America', 'Africa']
filtered_data = happy_df[happy_df['regional_indicator'].isin(regions_of_interest)]

# Calculate the mean of the salaries in the filtered data
mean_score_of_interest = filtered_data['happiness_score'].mean()

print("Mean happiness score for the regions of interest:", mean_score_of_interest)


# Specify the hypothesized population mean 
hypothesized_mean = 6.391310

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(filtered_data['happiness_score'], hypothesized_mean)

# Define the confidence level (in this case, 95%)
confidence_level = 0.95

# Calculate the critical t-value for the specified confidence level and degrees of freedom
degrees_of_freedom = len(filtered_data) - 1
critical_t = stats.t.ppf(confidence_level, df=degrees_of_freedom)

# Determine if the mean is at a 95% confidence level
is_confident = abs(t_stat) <= critical_t

print("Calculated Mean Score:", mean_score_of_interest)
print("Hypothesized Mean:", hypothesized_mean)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
print("Critical T-Value for 95% Confidence:", critical_t)
print("Is European Mean at 95% Confidence Level?", is_confident)
```

```python
# lets grab the GDP score by each region

mean_gdp = happy_df.groupby('regional_indicator')[['gdp_per_capita']].mean()
mean_gdp = mean_gdp.sort_values(by=['gdp_per_capita'], ascending=False)
mean_gdp
```

```python
# getting the mean of all other regions aside Asia to test the hypothesis
data = {
    'regional_indicator': ['Oceania', 'North America', 'South America', 'Asia', 'Africa', 'Europe'],
    'gdp_per_capita': [10.741500, 10.444333,9.748167, 9.478618, 9.197000, 8.196108]
}
happy_df = pd.DataFrame(data)

# Filter data for the specific months
gdp_of_interest = ['Oceania', 'North America', 'Europe', 'South America', 'Africa']
filtered_data = happy_df[happy_df['regional_indicator'].isin(gdp_of_interest)]

# Calculate the mean of the salaries in the filtered data
mean_score_of_interest = filtered_data['gdp_per_capita'].mean()

print("Mean gdp score for the regions of interest:", mean_score_of_interest)


# Specify the hypothesized population mean 
hypothesized_mean = 9.478618

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(filtered_data['gdp_per_capita'], hypothesized_mean)

# Define the confidence level (in this case, 95%)
confidence_level = 0.90

# Calculate the critical t-value for the specified confidence level and degrees of freedom
degrees_of_freedom = len(filtered_data) - 1
critical_t = stats.t.ppf(confidence_level, df=degrees_of_freedom)

# Determine if the mean is at a 95% confidence level
is_confident = abs(t_stat) <= critical_t

print("Calculated Mean GDP Score:", mean_score_of_interest)
print("Hypothesized Mean:", hypothesized_mean)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
print("Critical T-Value for 90% Confidence:", critical_t)
print("Is Asian GDP Mean at 90% Confidence Level?", is_confident)
```
